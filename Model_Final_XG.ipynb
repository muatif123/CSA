```python
%%time
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv("100k_survey_responses.csv")
df.head()
```

```python
df.isnull().value_counts()
```

```python
df.dropna(inplace=True)
df
```

```python
# Map single-choice questions
q1_map = {'Once a week': 1, '2-3 times a week': 2, 'Once a month': 3, 'Rarely': 4}
q3_map = {'₹100–₹500': 1, '₹500–₹1000': 2, '₹1000–₹2000': 3, '₹2000+': 4}

df['Q1'] = df['Q1'].map(q1_map)
df['Q3'] = df['Q3'].map(q3_map)
df['Q4'] = df['Q4'].astype(int)
df['Q8'] = df['Q8'].astype(int)
df['Q9'] = df['Q9'].astype(int)
df['Q10'] = df['Q10'].astype(int)
```

```python
df
```

```python
df_model = df[['Q1', 'Q3', 'Q4', 'Q8', 'Q9', 'Q10','Cust_satisfaction']]
df_model.head()
```

```python
df_model.isnull().sum()
```

```python
df_model.columns[df_model.columns.duplicated()]
```

```python
df_model = df_model.loc[:, ~df_model.columns.duplicated()]
df_model.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from xgboost import XGBClassifier
```

#### XGBoost Classifier

```python
# Train/test split
X = df_model.drop(columns=['Cust_satisfaction'])
y = df_model['Cust_satisfaction']
```

```python
# Add random noise to numeric columns
numeric_cols = X.select_dtypes(include=[np.number]).columns

# Add small Gaussian noise (mean=0, std=0.01 * std of each column)
noise = pd.DataFrame(
    np.random.normal(loc=0, scale=X[numeric_cols].std() * 0.01, size=X[numeric_cols].shape),
    columns=numeric_cols
)

# Add noise only to numeric features
X[numeric_cols] = X[numeric_cols] + noise
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
%%time
# Train RandomForest classifier
rf = RandomForestClassifier(random_state=123)
rf.fit(X_train, y_train)

xg = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xg.fit(X_train, y_train)

# Step 10: Evaluate
y_pred = xg.predict(X_test)

print("Accuracy:", xg.score(X_test, y_test))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import joblib

# Save model
joblib.dump(rf, 'customer_classifier_XG.pkl')
```
